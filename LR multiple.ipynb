# Linear Regression with multiple features

## Importing the libraries to work with the data

```python
import pandas as pd
import seaborn as sns
```

## Importing the data

```python
path="happy.csv"
data=pd.read_csv(path)
```

## Geting familiar with the data

```python
data.info()
```

Visualizing the correlation

```python
sns.heatmap(data=data.corr(), annot=True)
```

Let's predict the *Happiness Score* using the features *Economy*, *Family*, *Health*, and *Freedom*

Let's do a visualization

```python
sns.scatterplot(data=data[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom']])
```

## Declating the *features* and *targets*

Features

```python
features=['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom']
X=data[features]
X.head()
```

Target

```python
y=data["Happiness Score"]
y.head()
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8, test_size=0.2, random_state=0)
```

## Declaring the model

```python
from sklearn.linear_model import LinearRegression
LRmodel=LinearRegression()
```

## Fitting the model

```python
fit=LRmodel.fit(X_train,y_train)
```

Getting the values ${\theta}_{i}$ where $i=0,1,2,3,4$

Value ${\theta}_{0}$

```python
LRmodel.intercept_
```

Values ${\theta}_{i}$ where $i=1,2,3,4$

```python
LRmodel.coef_
```

Then we can see that ${\theta}_{4}$ which is `Freedom`  has more weight in out prediction

The model is 

$$
\hat{y}\left({x}_{1},{x}_{2},{x}_{3},{x}_{4}\right)={\theta}_{0}+\sum_{i=1}^{4}\left({x}_{i}{\theta}_{i}\right)
$$
